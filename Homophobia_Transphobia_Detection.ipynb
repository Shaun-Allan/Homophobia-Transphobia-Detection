```python
import pandas as pd
import numpy as np
import string
import re
import nltk
from nltk.corpus import stopwords
#For Feature Extraction
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
#ML Algorithms - Naive Bayes, XGBoost, SVM
from sklearn.naive_bayes import MultinomialNB
from xgboost import XGBClassifier
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
#Transformer models
#To check the performance of these models
from sklearn.metrics import classification_report
import gensim
from gensim.parsing.preprocessing import STOPWORDS
from collections import Counter
```

```python
label={"Non-anti-LGBT+ content":0,"Homophobia":1,"Transphobia":2}
```

```python
def tolower(text):
  return text.lower()
```

```python
def removepunctuation(text):
    clean="".join([i for i in text if i not in string.punctuation])
    return clean
```

```python
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
```

```python
def preprocess(df):
  df['text']= df['text'].apply(lambda x:tolower(x))
  df['text']= df['text'].apply(lambda x:removepunctuation(x))
  df['text']= df['text'].apply(lambda x:deEmojify(x))
  df['category'] = df['category'].map(label)
  return df
```

```python
from simpletransformers.classification import ClassificationModel
```

```python
def train_bert_model(model,df):
    bert = ClassificationModel('bert', model, num_labels=8, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 1,
    # 'silent': True
    })
    bert.train_model(df)
    return bert
```

```python
def find_accuracy(bert,df):
    result, model_outputs, wrong_predictions = bert.eval_model(df)

    predictions, raw_outputs = bert.predict(df['text'].tolist())
    print("predictions:")
    print(predictions)

    bads = {}
    for pred in wrong_predictions:
        if pred.label in bads:
            bads[pred.label] += 1
        else:
            bads[pred.label] = 1
    print("wrong predictions:")
    wrong = len(wrong_predictions)
    total = len(df)
    print(str(wrong) + ' wrong out of ' + str(total))
    print(bads)
    accuracy = (total-wrong)/total*100
    print("Accuracy: "+str(accuracy)+"%")

    print("Classification report: ")
    from sklearn.metrics import classification_report
    print(classification_report(df.category, predictions))
```

```python
model=train_bert_model("l3cube-pune/malayalam-bert",malayalam)
```

```python
find_accuracy(model,malayalam_train)
```

```python
model.train_model(malayalam_dev_train)
```

```python
find_accuracy(model,malayalam_dev_train)
```
